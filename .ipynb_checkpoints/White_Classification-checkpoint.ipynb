## Data Cleaning



```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy import stats
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor

from pandas.plotting import scatter_matrix
from sklearn.preprocessing import QuantileTransformer
from pandas import DataFrame as df
```

```python
#Read in csv file
white_ca = pd.read_csv("White_Analysis/winequality_white.csv", sep=';', encoding='UTF8') 
# Drop the null columns where all values are null
white_ca = white_ca.dropna(axis='columns', how='all')
# Drop the null rows
white_ca.dropna()
white_ca.head()
white_ca
```

```python
for q in white_ca['qualtiy']:
    if q > 5:
        white_ca['qualtiy'] =  white_ca['qualtiy'].replace([6, 7, 8,9,10], 1)
    else:
        white_ca['qualtiy'] = white_ca['qualtiy'].replace([1,2,3, 4, 5], 0)
```
